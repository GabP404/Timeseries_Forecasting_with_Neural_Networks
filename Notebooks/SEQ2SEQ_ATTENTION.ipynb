```python
def build_sequences(train=True, window=200, stride=20, telescope=100):
    # Sanity check to avoid runtime errors
    assert window % stride == 0
    dataset = []
    labels = []

    maxIndex = train_data.shape[0] if train==True else test_data.shape[0]

    for index in range(maxIndex):

      if train == True:
        sequence = getTrainSequenceFromDataset(index)
        temp_sequence = getTrainSequenceFromDataset(index).copy()
      else:
        sequence = getTestSequenceFromDataset(index)
        temp_sequence = getTestSequenceFromDataset(index).copy()

      temp_label = sequence.copy()

      padding_check = len(sequence)%window

      if(padding_check != 0):
        # Compute padding length
        padding_len = window - len(sequence)%window
        padding = np.zeros((padding_len), dtype='float32')
        temp_sequence = np.concatenate((padding,sequence))
        padding = np.zeros((padding_len), dtype='float32')
        temp_label = np.concatenate((padding,temp_label))
        assert len(temp_sequence) % window == 0

      for idx in np.arange(0,len(temp_sequence)-window-telescope,stride):
        dataset.append(temp_sequence[idx:idx+window])
        labels.append(temp_label[idx+window:idx+window+telescope])

      del sequence
      del temp_sequence
      del temp_label

    dataset = np.array(dataset)
    labels = np.array(labels)
    return dataset, labels
```

```python
def attention_layer(encoder_outputs, decoder_hidden):
  """
  Attention layer to compute attention weights.

  Args:
      encoder_outputs: Stacked LSTM outputs from the encoder (shape: (batch, timesteps, units))
      decoder_hidden: Hidden state of the decoder LSTM (shape: (batch, 200, units))  # Adjusted shape

  Returns:
      A weighted context vector (shape: (batch, units)) and attention weights (shape: (batch, timesteps))
  """
  # Attention mechanism (Bahdanau attention)
  score = tfkl.Dot(axes=1)([decoder_hidden, encoder_outputs])  # Dot product on last two axes
  attention_weights = tfkl.Softmax(axis=1)(score)
  context_vector = tfkl.Dot(axes=1)([attention_weights, encoder_outputs])
  return context_vector, attention_weights




def build_model(timesteps, n_features, n_predictions, units=32):
    # Define encoder
    encoder_inputs = tfk.Input(shape=(timesteps, n_features))
    encoder_lstm = tfkl.LSTM(units=units, return_sequences=True)
    encoded_outputs = encoder_lstm(encoder_inputs)

    # Define decoder
    decoder_dense = tfkl.Dense(units=n_features)  # Removed TimeDistributed
    decoder_hidden = decoder_dense(encoded_outputs)

    # Reshape decoder_hidden to remove timestep dimension
    decoder_hidden = tfkl.Flatten()(decoder_hidden) 

    # Attention layer
    context_vector, attention_weights = attention_layer(encoded_outputs, decoder_hidden)

    # Decoder with attention
    decoder_combined = tfkl.Concatenate(axis=-1)([decoder_hidden, context_vector])
    decoder_dense = tfkl.Dense(units=n_features)(decoder_combined)

    # Build model
    model = tfk.Model(inputs=encoder_inputs, outputs=decoder_dense)

    # Compile model
    model.compile(loss='mse', optimizer='adam')

    return model

```

```python
from tensorflow import keras
from tensorflow.keras.layers import LSTM, Dense, Input, Attention, Concatenate
from tensorflow.keras.models import Model

def build_model(timesteps_in, timesteps_out, n_features):
    """
    Builds a seq2seq model with Attention layer.

    Args:
        timesteps_in: Number of input timesteps.
        timesteps_out: Number of output timesteps.
        n_features: Number of features.

    Returns:
        A Keras Model instance.
    """

    # Encoder
    encoder_inputs = Input(shape=(timesteps_in, n_features))
    encoder_lstm = LSTM(128, return_sequences=True)(encoder_inputs)
    encoder_outputs, state_h, state_c = LSTM(128, return_state=True)(encoder_lstm)
    encoder_states = [state_h, state_c]

    # Decoder
    decoder_inputs = Input(shape=(timesteps_out, n_features))
    decoder_lstm = LSTM(128, return_sequences=True)(decoder_inputs, initial_state=encoder_states)

    # Attention (Using Attention layer)
    attention = Attention(use_scale=True)([decoder_lstm, tfkl.RepeatVector(timesteps_out)(encoder_outputs)])
    decoder_combined_context = Concatenate(axis=-1)([attention, decoder_lstm])

    # Output
    decoder_outputs = Dense(n_features)(decoder_combined_context)

    # Model
    model = Model(inputs=[encoder_inputs, decoder_inputs], outputs=decoder_outputs)
    model.compile(loss=tfk.losses.MeanSquaredError(), 
                  optimizer=tfk.optimizers.Adam(), 
                  metrics=['mae'])

    return model
```

```python

def predict_sequence(model, X_test, y_test):
    predictions = []
    decoder_input_data = np.zeros((y_test.shape[0], y_test.shape[1], 1))

    for i in range(y_test.shape[1]):
        # Predict the output for the current timestep
        output = model.predict([X_test, decoder_input_data])

        # Update the decoder input with the predicted output
        decoder_input_data[:, i, :] = output[:, i, :]

        # Store the prediction
        predictions.append(output[:, i, :])

    # Concatenate predictions into a single array
    predictions = np.concatenate(predictions, axis=1)

    return predictions
```
